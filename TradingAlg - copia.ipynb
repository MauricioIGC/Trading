Como ya se mencionó la idea es que el precio de los activos sigue las leyes de Newton, por eso haré un algoritmo bottom up que haga regresiones lineales y divida a la mitad los datos hasta que el error sea ínfimo

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

def bottom_up_segmentation(series, max_error, min_segment_length):
    segments = [(0, len(series) - 1)]
    auxiliar_error=[]
    while True:
        max_segment_error = 0
        best_segment = None
        for (start, end) in segments:
            if end - start > min_segment_length:
                segment = series[start:end + 1]
                x = np.arange(len(segment)).reshape(-1, 1)
                y = segment.values
                model = LinearRegression().fit(x, y)
                predictions = model.predict(x)
                error = np.sum((y - predictions) ** 2)
                if error > max_segment_error:
                    auxiliar_error.append(error)
                    max_segment_error = error
                    best_segment = (start, end)
        if max_segment_error < max_error:
            break
        segments.remove(best_segment)
        mid = (best_segment[0] + best_segment[1]) // 2
        segments.append((best_segment[0], mid))
        segments.append((mid + 1, best_segment[1]))
    return segments, auxiliar_error

# Ejemplo de uso
series = ticks_frame["ask"]
segments, auxiliar_error = bottom_up_segmentation(series, max_error=.02, min_segment_length=10)
print(segments)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

def bottom_up_segmentation_3(series, max_error, min_segment_length):
    segments = [(0, len(series) - 1)]
    auxiliar_error=[]
    while True:
        max_segment_error = 0
        best_segment = None
        for (start, end) in segments:
            if end - start > min_segment_length:
                segment = series[start:end + 1]
                x = np.arange(len(segment)).reshape(-1, 1)
                y = segment.values
                model = LinearRegression().fit(x, y)
                predictions = model.predict(x)
                error = np.sum((y - predictions) ** 2)
                if error > max_segment_error:
                    auxiliar_error.append(error)
                    max_segment_error = error
                    best_segment = (start, end)
        if max_segment_error < max_error:
            break
        segments.remove(best_segment)
        mid1 = best_segment[0] + (best_segment[1] - best_segment[0]) // 3
        mid2 = best_segment[0] + 2 * (best_segment[1] - best_segment[0]) // 3
        segments.append((best_segment[0], mid1))
        segments.append((mid1 + 1, mid2))
        segments.append((mid2 + 1, best_segment[1]))
    return segments, auxiliar_error

# Ejemplo de uso
series = ticks_frame["ask"]
segments, auxiliar_error = bottom_up_segmentation_3(series, max_error=.02, min_segment_length=10)
print(segments)
```
